```python
null_var = df.isnull().sum()/df.shape[0]*100
print(null_var)
```

```python
drop_column = null_var[null_var>17]
print(drop_column)
```

```python
drop_column = df.drop(columns = drop_column.index)
```

```python
drop_column.shape
```

```python
df.shape
```

```python
df1 = drop_column.select_dtypes(include = ['int64','float64'])
```

```python
df1.head()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df1.isnull())
```

```python
df1[df1.isnull().any(axis = 1)]
```

```python
df1.isnull().sum()
```

```python
missing_num_var = [var for var in df1.columns if df1[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(20,20))
sns.set()
for i, var in enumerate(missing_num_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df1[var],bins=20,kde_kws={'linewidth':5, 'color':'#DC143C'})
```

```python
df_num_mean = df1.fillna(df1.mean())
df_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df1[var],bins=20,kde_kws={'linewidth':8, 'color':'red'},label = "Orignal")
  sns.distplot(df_num_mean[var],bins=20,kde_kws={'linewidth':5, 'color':'green'}, label = "Mean")
  plt.legend()
```

```python
df_num_median = df1.fillna(df1.median())
df_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(20,20))
sns.set()
for i, var in enumerate(missing_num_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df1[var],bins=20,kde_kws={'linewidth':8, 'color':'red'},label = "Orignal")
  sns.distplot(df_num_mean[var],bins=20,kde_kws={'linewidth':5, 'color':'green'}, label = "Mean")
  sns.distplot(df_num_median[var],bins=20,kde_kws={'linewidth':5, 'color':'yellow'}, label = "Median")
  plt.legend()
```

```python
for i, var in enumerate(missing_num_var):
  plt.figure(figsize=(10,10))
  plt.subplot(2,2,i+1)
  sns.boxplot(df1[var])
  plt.subplot(2,2,i+2)
  sns.boxplot(df_num_mean[var])
  plt.subplot(2,2,i+3)
  sns.boxplot(df_num_median[var])
```

```python
df_concat = pd.concat([df1[missing_num_var],df_num_mean[missing_num_var],df_num_median[missing_num_var]])
```
