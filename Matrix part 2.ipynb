```python
x = 5
y = np.random.randint(100,size=(3,3))
print("Your Matrix is :\n",y)
print("Your Scalar is :\n",x)
print("Multiplication of Matrix and Scalar is :")
print(x*y)
```

**Multiplying Two Matrix**

Rule = No of column in first matrix is equal to the no rows in the second Matrix and the resultant Matrix will have same no of column in first matrix and the same no of rows at the second matrix  

```python
a = np.random.randint(100,size=(3,3))
b= np.random.randint(100, size=(3,3))
print("First Matrix \n",a)
print()
print("Second Matrix \n" ,b)
print()
c = np.dot(a,b)
print("Multiplication of Two Matrix \n ",c)
```
