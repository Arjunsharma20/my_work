Numpy is a low level library written in C (and FORTRAN) for high level mathematical function . Numpy cleverly overcomes the problem of running slower algorithim on Python uysing multidimentional arrats and function that operate on array. Any algorithim can then be epressed as a function on arrays,allowing the algorithim to be run quickly.

```python
import numpy as np
```

```python
x = np.array([1,2,3])
print(x)
```

```python
y = np.array([[[1,2,3],[5,6,7]],[[7,8,9],[9,5,3]]])
print(y)
```

```python
# using numpy function
z = np.zeros((3,3))
print(z)
```

```python
a = np.ones((3,3))
print(a)
```
