```python
df[df.loc[:,'LotConfig'] == "Inside"]['LotFrontage'].replace(np.nan,df[df.loc[:,'LotConfig'] == "Inside"]['LotFrontage'].mean())
# ye only inside class me mean value dalegay
```

replace the null value of lotfrontage we use loop

```python
# ye hr class me mean value dalega
df_num_copy = df_num.copy()  #first we make a copy of our dataframe
for var_class in df['LotConfig'].unique():
# ye lotconfig ki hr class ki value pr jaega ['Inside', 'FR2', 'Corner', 'CulDSac', 'FR3']
    df_num_copy.update(df[df.loc[:,'LotConfig'] == var_class]['LotFrontage'].replace(np.nan,df[df.loc[:,'LotConfig'] == var_class]['LotFrontage'].mean()))
    #  df_copy.update ye update krega hmare df copy dataframe ko
    # df[df.loc[:,'LotConfig'] == var_class] ye locate krega hme Lotconfig ki first class me
    # ['LotFrontage'].replace(np.nan,df[df.loc[:,'LotConfig'] == var_class]['LotFrontage'].mean() ye uss jgh pr nan value ko replace krdega mean se
```

```python
df_num_copy.isnull().sum()
```

```python
df_copy = df.copy()
num_vars_miss = ['LotFrontage','MasVnrArea','GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_vars, num_vars_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_vars].unique():
        df_copy.update(df[df.loc[:,cat_vars] == var_class][num_vars_miss].replace(np.nan,df[df.loc[:,cat_vars] == var_class][num_vars_miss].mean()))
```

```python
df_copy.isnull().sum()
```

Data Distribution

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
  plt.subplot(2,2,i+1)
  sns.distplot(df[var],bins=20,kde_kws={'linewidth':8, 'color':'red'},label = "Orignal")
  sns.distplot(df_copy[var],bins=20,kde_kws={'linewidth':5, 'color':'green'}, label = "Mean")
  plt.legend()
```

```python
df_median_copy = df.copy()
num_vars_miss = ['LotFrontage','MasVnrArea','GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_vars, num_vars_miss in zip(cat_vars,num_vars_miss):
    for var_class in df[cat_vars].unique():
        df_median_copy.update(df[df.loc[:,cat_vars] == var_class][num_vars_miss].replace(np.nan,df[df.loc[:,cat_vars] == var_class][num_vars_miss].median()))
```

```python
df_median_copy[num_vars_miss].isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_vars_miss):
  plt.subplot(2,2,i+1)
  sns.distplot(df[var], bins=20,kde_kws={'linewidth':8, 'color':'red'},label = "Orignal")
  sns.distplot(df_copy[var],bins=20,kde_kws={'linewidth':5, 'color':'green'}, label = "Mean")
  sns.distplot(df_median_copy[var],bins=20,kde_kws={'linewidth':3, 'color':'k'}, label = "Median")
  plt.legend()
```
