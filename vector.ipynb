```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
# gradient dasent loss function
plt.quiver(5,1,2,5)
plt.show()
```

```python
plt.quiver(5,1,2,5,angles='xy',scale_units='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(5,1,2,5,angles='xy',scale_units='xy',scale=1)
plt.quiver(5,1,-2,-5,angles='xy',scale_units='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Adding Two Vector

```python
vector_1 = np.asarray([5,1,2,3])
vector_2 = np.asarray([2,5,3,-2])
sum =vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,5,1,angles='xy',scale_units='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two vwctor


```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
sub =vector_1 - vector_2
print(sub)
```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,-1,5,angles='xy',scale_units='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiply a Vector by a scaler

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = 2*vector_1
# mul =vector_1 * vector_2
print(vector_2)
```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1,color='b')
# plt.quiver(0,0,3,-2,angles='xy',scale_units='xy',scale=1,color='r')
plt.quiver(0,0,4,6,angles='xy',scale_units='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = -0.5*vector_1
# mul =vector_1 * vector_2
print(vector_2)
```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xyx',scale=1,color='b')
plt.quiver(0,0,-1,-1.5,angles='xy',scale_units='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
