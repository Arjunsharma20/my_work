```python
import numpy as np
```

Dot Product of 2 Vector

```python
a = np.array([2,3])
b = np.array([4,4])
c = np.dot(a, b)
```

```python
print(c)
```

```python
d = np.array([40,30,25])
e = np.array([53,24,68])
f= np.dot(d, e)
print(f)
```

Cross Product of 2 Vector

```python
a = np.array([2,3])
b = np.array([4,4])
c = np.cross(a, b)
print(c)
```

```python
d = np.array([40,30,25])
e = np.array([53,24,68])
f= np.cross(d, e)
print(f)
```

Projection of a Vector on b vector
images.png

```python
a = np.array([2,5])
b = np.array([8,-6])
# Magnitude of b vector
magnitude_of_b= np.sqrt(sum(b**2))
# c = np.dot(a, b)/np.dot(b, b) * b
proj_of_a_on_b = (np.dot(a,b)/magnitude_of_b**2 )*b
print(proj_of_a_on_b)
# print(c)
```
