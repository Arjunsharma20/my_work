```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for ours plote
import numpy as np
```

```python
# making a numpy array
x=np.array([1,5,3,9,10])
y=np.array([10,50,30,90,10])
```

```python
print(x)
```

```python
# using font function to set font properties
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'pink','size':15}
```

```python
#give label and title
plt.title("Result",font1)
plt.xlabel("Average",font2)
plt.ylabel("Total",font2)
plt.plot(x,y)
```

```python
# parabola
a=np.linspace(-10,10,20)
b=a**2
plt.bar(a,b)
plt.show
```

```python
plt.plot(a,b,'r+')
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

```python
#using bar graph
x=np.array(["Arjun","Ishitiv","Sakshi","Biswajit","Akshay"])
y=np.array([90,95,25,88,90])
plt.xlabel("Student Name")
plt.ylabel("Percentage")
plt.title("Result")
plt.bar(x,y)
```

```python
# usning piechart
y=np.array([90,95,25,88,90])
mylabel=(["Arjun","Ishitiv","Sakshi","Biswajit","Akshay"])
myexplode=[0.2,0,0,0,0]
plt.pie(y,labels = mylabel ,explode = myexplode ,shadow=True)
plt.legend(title="Student Name:")
```
