                   **Gradient Descent**
Gradient Descent is an optimization algorithim used for minimizing the cost function in various machine learning algorithim It is used fir updating the parameters of the learning model.

w = w - l*dw

b = b -l*db

**Learning Rate:**

Learning Rate is a tuning parameter in an optimization algorithim that determinrs the step size at each iteration while moving toward a minimum of a loss function.

                         or

The learning rate is a hyperparameter that determines the step size during the optimization process (gradient descent) of a machine learning model.

Derivatives



**dw = 1/m * (Y(pred) - Y).X**

**db = 1/m * (Y(pred) - Y)**


```python
#  Import the Dependencies
import numpy as np
```

**Logistic Regression**

```python
class LogisticRegression():
  # Declaring the learning rate and the number of iteration (Hyperparameter)
  def __init__(self,learning_rate,no_of_iteration):
    self.learning_rate = learning_rate
    self.no_of_iteration = no_of_iteration
# Fit function to train the model with some dataset
  def fit(self,X,Y):
    # number of rows and column in the dataset
    self.m,self.n = X.shape # m represent the total no of rows (dat point) and n represent the total no of column (input features)
    # Inatiating weight and bias value
    self.w = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y

    # Implementing Gradient Descent for optimization
    for i in range(self.no_of_iteration):
      self.update_weight(self.Y)


  def update_weight(self,Y):

    # Y(pred) formula known as sigmoid function
    Y_pred = 1 / (1 + np.exp(-(self.X.dot(self.w)+ self.b)))    # wX + b
    # Find the Derivatives
    dw = (1/self.m) * np.dot(self.X.T,(Y_pred - self.Y))
    db = (1/self.m) * np.sum(Y_pred - self.Y)

    #  updating the weight and bias using Gradient Descent equation
    self.w = self.w - self.learning_rate * dw
    self.b = self.b - self.learning_rate * db

    # Sigmoid Equation and Decision Boundary

  def predict(self,X):
    Y_prediction = 1 / (1 + np.exp(-(self.X.dot(self.w)+ self.b)))
    Y_prediction = np.where(Y_prediction >0.5 , 1 , 0)
    return Y_prediction
```
