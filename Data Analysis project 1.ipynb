**Exploratory Data Analysis (EDA)**

Gender


```python
df.columns
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(data = df,x = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending =False)
x='Gender'
y='Amount'
plt.bar(x,y,data=sales_gen)
```

from the above graph we can see that most of the buyer are females or even the purchasing power of female is greater than men

Age

```python
df.columns
```

```python
plt.figure(figsize=(5,4))
ax = sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

from above graph we can see that the age droup of 26 to 35 spend more money
