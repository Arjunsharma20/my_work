```python
import numpy as np
```

Creating a Matrix using Numpy

```python
matrix_1= np.array([[1,7],[2,0]])
print(matrix_1)
```

```python
matrix_1.shape
```

```python
matrix_2 = np.array([[1,2,0],[7,0,0],[5,6,2]])
print(matrix_2)
```

```python
matrix_2.shape
```

Creating Matrix with Random values

```python
random_matrix= np.random.rand(7,9)
print(random_matrix)
```

```python
random_integers_matrix = np.random.randint(100, size=(8,9))
print(random_integers_matrix)
```

Matrix with all value one

```python
matrix_3=np.ones((3,4),dtype=int)
print(matrix_3)
```

Matrix with all value zero

```python
matrix_4 = np.zeros((5,5),dtype=int)
print(matrix_4)
```

Identity Matrix

```python
identity_matrix = np.eye(7,9,dtype=int)
print(identity_matrix)
```

Transpose Matrix

```python
a = np.random.randint(100, size=(5,6))
print(a)
```

```python
Transpose_a= np.transpose(a)
print(Transpose_a)
```
